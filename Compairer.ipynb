```python
import warnings 
warnings.filterwarnings('ignore') 
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read book1.xlsx
book1 = pd.read_excel('book1.xlsx')
book1.head()

```

```python
#read book2.xlsx
book2 = pd.read_excel('book2.xlsx')
book2.head()
```

```python
#rename columns of book2 Power to New_Power 
book2.rename(columns={'Power':'New_Power'}, inplace=True)
#rename columns of book2 T1-kills to New_T1kills
book2.rename(columns={'T1-Kills':'New_T1kills'}, inplace=True)
#rename columns of book2 T2-kills to New_T2kills
book2.rename(columns={'T2-Kills':'New_T2kills'}, inplace=True)
#rename columns of book2 T3-kills to New_T3kills 
book2.rename(columns={'T3-Kills':'New_T3kills'}, inplace=True)
#rename columns of book2 T4-kills to New_T4kills
book2.rename(columns={'T4-Kills':'New_T4kills'}, inplace=True)
#rename columns of book2 T5-kills to New_T5kills
book2.rename(columns={'T5-Kills':'New_T5kills'}, inplace=True)
#rename columns of book2 Dead Troops to New_Dead_Troops
book2.rename(columns={'Dead Troops':'New_Dead_Troops'}, inplace=True)
#rename columns of book2 rss assistance to New_rss_assistance
book2.rename(columns={'Rss Assistance':'New_rss_assistance'}, inplace=True)
#rename columns of book2 Alliance Helps to New_Alliance_Helps
book2.rename(columns={'Alliance Helps':'New_Alliance_Helps'}, inplace=True)
#rename columns of book2 Rss gathered to New_Rss_gathered   
book2.rename(columns={'Rss Gathered':'New_Rss_gathered'}, inplace=True)
book2.head()
```

```python
df = pd.merge(book1, 
                     book2, 
                     on ='Name', 
                     how ='left')
df
```

```python
#drop Alliance_y 
df.drop(['Alliance_y'], axis=1, inplace=True)
#rename Alliance_x to Alliance
df.rename(columns={'Alliance_x':'Alliance'}, inplace=True)
```

```python
#create new column difference between New_Power and Power
df['difference_power'] = df['New_Power'] - df['Power']
#create new column difference_t1 between New_T1kills and T1-Kills
df['difference_t1'] = df['New_T1kills'] - df['T1-Kills']
#create new column difference_t2 between New_T2kills and T2-Kills
df['difference_t2'] = df['New_T2kills'] - df['T2-Kills']
#create new column difference_t3 between New_T3kills and T3-Kills
df['difference_t3'] = df['New_T3kills'] - df['T3-Kills']
#create new column difference_t4 between New_T4kills and T4-Kills
df['difference_t4'] = df['New_T4kills'] - df['T4-Kills']
#create new column difference_t5 between New_T5kills and T5-Kills
df['difference_t5'] = df['New_T5kills'] - df['T5-Kills']
#create new column difference_dead_troops between New_Dead_Troops and Dead Troops
df['difference_dead_troops'] = df['New_Dead_Troops'] - df['Dead Troops']
#create new column difference_rss_assistance between New_rss_assistance and Rss Assistance
df['difference_rss_assistance'] = df['New_rss_assistance'] - df['Rss Assistance']
#create new column difference_alliance_helps between New_Alliance_Helps and Alliance Helps
df['difference_alliance_helps'] = df['New_Alliance_Helps'] - df['Alliance Helps']
#create new column difference_rss_gathered between New_Rss_gathered and Rss Gathered
df['difference_rss_gathered'] = df['New_Rss_gathered'] - df['Rss Gathered']
df.head()


```

```python
#create variable difference_power_percentage
df['difference_power_percentage'] = df['difference_power'] / df['Power'] * 100
#create variable difference_t1_percentage
df['difference_t1_percentage'] = df['difference_t1'] / df['T1-Kills'] * 100
#create variable difference_t2_percentage
df['difference_t2_percentage'] = df['difference_t2'] / df['T2-Kills'] * 100
#create variable difference_t3_percentage
df['difference_t3_percentage'] = df['difference_t3'] / df['T3-Kills'] * 100
#create variable difference_t4_percentage
df['difference_t4_percentage'] = df['difference_t4'] / df['T4-Kills'] * 100
#create variable difference_t5_percentage
df['difference_t5_percentage'] = df['difference_t5'] / df['T5-Kills'] * 100
#create variable difference_dead_troops_percentage
df['difference_dead_troops_percentage'] = df['difference_dead_troops'] / df['Dead Troops'] * 100
#create variable difference_rss_assistance_percentage
df['difference_rss_assistance_percentage'] = df['difference_rss_assistance'] / df['Rss Assistance'] * 100
#create variable difference_alliance_helps_percentage
df['difference_alliance_helps_percentage'] = df['difference_alliance_helps'] / df['Alliance Helps'] * 100
#create variable difference_rss_gathered_percentage
df['difference_rss_gathered_percentage'] = df['difference_rss_gathered'] / df['Rss Gathered'] * 100
df.head()




```
